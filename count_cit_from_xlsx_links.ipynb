```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.chrome.options import Options
import re
import pandas as pd
```

Чтение исходного файла

```python
df = pd.read_excel("tabl_it.xlsx")
df
```

Заполнение столбцов по годам нулями

```python
if "2000" not in df:
    for i in range(1980, 2026):
        df[str(i)] = 0
df
```

Проход по сайтом и сбор данных (кол-во цитирований по годам)

```python
options = Options()
options.page_load_strategy = 'none'
driver = webdriver.Chrome(options=options)
sch = 0
dict_uniq_links = {}
for i in df['link']:
    if i in dict_uniq_links:
        df.loc[sch, "flag_checked"] = 1
        for jk in range(1980, 2025):
            df.loc[sch, jk] = int(dict_uniq_links[i][jk])
    if df.loc[sch, "flag_checked"] == 1:
        sch += 1
        continue
    sites = i
    driver.get(sites)
    time.sleep(11)
    list_citir_links_in_art = driver.find_elements(By.CLASS_NAME, "relative-size-cont-number")
    for j in list_citir_links_in_art:
        clck = j
        # print('bebeb')
        # print(clck.get_attribute("innerHTML"))
        if j.get_attribute("innerHTML").replace('\n', '').replace(" ", '').find('href="#"') != -1:
            time.sleep(1)
            clck.click()
            cnt = j.text
            time.sleep(5)
            break
        time.sleep(1)
    # print("_________", i)
    time.sleep(6)
    citir_year_text = driver.find_element(By.ID, "framed_div").get_attribute("innerHTML")
    a1 = [j.start() for j in re.finditer("<strong>", citir_year_text)]
    a2 = [j.start() for j in re.finditer("</strong>", citir_year_text)]
    d_year_freq = {pz : 0 for pz in range(1980, 2026)}
    for j in range(len(a1)):
        d_year_freq[int(citir_year_text[a1[j] + 8:a2[j]])] += 1
    # print(d_year_freq)
    df.loc[sch, "flag_checked"] = 1
    dict_uniq_links[i] = d_year_freq
    for jk in range(1980, 2026):
        df.loc[sch, jk] = int(dict_uniq_links[i][jk])
    sch += 1
```
